```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


#Read the data fron csv file

data = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_5795797.csv")

#Now plotting bar chart for categorical variable

count_of_gender = data['Region'].value_counts()

bar_width = 0.5
x = range(len(count_of_gender.index))

plt.bar(count_of_gender.index , count_of_gender.values)

#Detailing all then necessary labels
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Region Distribution')

plt.xticks(x, count_of_gender.index , rotation=45)

for i , count in enumerate(count_of_gender.values):
    plt.text(i , count , str(count) , ha='center',va='bottom')

plt.tight_layout()
plt.show()


```
